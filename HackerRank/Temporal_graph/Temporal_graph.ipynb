```python
import heapq, io

def run_temporal_graph(data):
    print(data)
    input_data = io.StringIO(data).readline
    
    N, M = map(int, input_data().split())
    S, T = map(int, input_data().split())
    
    adj = [[] for _ in range(N + 1)]
    print(adj)

    for _ in range(M):
        # print(input_dat)
        u, v, w, L, R = map(int, input_data().split())
        adj[u].append((v, w, L, R))

    print(adj)
    INF = 10**30
    dist = [INF] * (N + 1)
    dist[S] = 0

    pq = [(0, S)]
    
    while pq:
        t, u = heapq.heappop(pq)
        if t != dist[u]: 
            continue
        for v, w, L, R in adj[u]:
            if t > R:
                continue
            start_time = max(t, L)
            arrival = start_time + w
            if arrival < dist[v]:
                dist[v] = arrival
                heapq.heappush(pq, (arrival, v))

        return dist[T] if dist[T] < INF else "No path"
```

```python
test_input = """\
5 6
1 5
1 2 3 0 10
2 3 4 5 20
3 5 2 0 100
1 4 5 3 6
4 3 3 6 8
2 5 20 5 7
"""

print(run_temporal_graph(test_input))
```
