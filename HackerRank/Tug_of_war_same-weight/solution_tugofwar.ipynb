```python
# Input
N = int(input())
weights = [int(input()) for _ in range(N)]
total = sum(weights)

team1_size = N // 2
team2_size = N - team1_size

# Initialize DP table: dp[k][s] = True if we can pick k students with sum s
dp = [[False] * (total + 1) for _ in range(N + 1)]
dp[0][0] = True  # base case: 0 students give sum 0

# DP logic
for w in weights:
    for k in range(N - 1, -1, -1):
        for s in range(total - w, -1, -1):
            if dp[k][s]:
                dp[k + 1][s + w] = True

# Find best split
best = 0
for k in [team1_size, team2_size]:
    for s in range(total + 1):
        if dp[k][s] and abs(total - 2 * s) < abs(total - 2 * best):
            best = s

t1 = best
t2 = total - best
if t1 > t2:
    t1, t2 = t2, t1

print(t1, t2)
```
